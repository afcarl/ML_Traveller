# Predict the Missing Grade

For the machine learning task I decided to solve the [Predict the Missing Grade](https://www.hackerrank.com/challenges/predict-missing-grade) problem. My solution is currently ranked 9th on the leaderboard.

## Dataset

The goal is to predict a grade for the Mathematics course when given 4 grades of 4 other courses (English is always present). I noticed that Hindi has never any values in the training set, so I left out this course from the final model. 

The grade is a value between 1 and 8. This means that the data has an ordinal distribution. While I eventually approximated the ordinal distribution via a linear regression output, better results might be get by using a ordinal regression regression model to predict the grade (e.g. an ordered logit).

The dataset was examined by plotting the 2D distribution and printing the correlation for each course compared to Mathematics [see below]. It can be noted on these figures that the grades of the courses seem to be more or less linearly correlated with Mathematics, but that the factor of correlation is quite low.

## Model

The final model used was the [SGDRegressor](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html). The SGDRegressor is a linear model trained by stochastic gradient descent. It's loss function can be chosen so that performs linear support vector regression, this is done by choosing `epsilon_insensitive` as the loss function. Support vector regression is prefered over linear regression in that is often more robust, and it generalises wel over the data.

Nonlinear models were tested, but performed the same, worse, or had other difficulties. A random forest model seemed to get better results on the test set, but the model persisted to a string was to big to load into the hacker rank platform. Non-linear support vectors machines were to slow to train because the number of samples is quite high (+/- 80,000). Neural networks didn't gave better results either.

## Final grade

Because the scoring function used in this challenge allows a misclassification up to one grade in difference the final score is adjusted. All grades that are predicted to be less than 2 are set to 2, and all graded that are predicted to be greater than 7 are set to be 7. The bound on the grades are 1 and 8, so predicting 2 or 7 instead of 1 or 8 gives no difference in error.

```python
import json
import numpy as np
import scipy
import scipy.stats
import sklearn
import sklearn.linear_model
import itertools
import os
import zlib
import base64
import pickle

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Courses: 
# Create a list that holds the courses
course_list = [
    'Mathematics', # 0
    'English', # 1
    'Physics', # 2
    'Chemistry', # 3
    'ComputerScience', # 4
    'Hindi', # 5
    'Biology', # 6
    'PhysicalEducation', # 7
    'Economics', # 8
    'Accountancy', # 9
    'BusinessStudies' # 10
]
# Create dictionary to hold index values so that these can easily be found when needed.
index_dict = dict()
for i in xrange(len(course_list)):
    index_dict[course_list[i]] = i
    
number_of_courses = len(index_dict)

# Define the course columns to extract for the X matrix
X_columns = [1,2,3,4,6,7,8,9,10]

def get_json_courses(sample_line):
    """
    Function to extract a json object with grades from sample_line.
    Get each course from this json object and return as a vector
    """
    x = np.zeros((number_of_courses,), dtype=np.int)
    sample = json.loads(sample_line)
    for course, grade in sample.iteritems():
        if course != 'serial':
            x[index_dict[course]] = grade
    return x


def read_training_data():
    """
    Read the training data and return X, Y
    """
    train_file ='training-and-test/training.json'
    with open(train_file, 'r') as f:
        nb_of_samples = int(next(f))
        # create the array that holds X and Y
        XY = np.zeros((nb_of_samples, number_of_courses), dtype=np.int)
        for idx, line in enumerate(f):
            x = get_json_courses(line)
            XY[idx, :] = x
    # split XY in X and Y
    return XY[:,X_columns], XY[:,0]


def read_test_data():
    """
    Read the test data and return X, Y
    """
    test_x_file = 'training-and-test/sample-test.in.json'
    test_y_file = 'training-and-test/sample-test.out.json'
    with open(test_x_file, 'r') as fx:
        nb_of_samples = int(next(fx))
        # create the array that holds X and Y
        XY = np.zeros((nb_of_samples, number_of_courses), dtype=np.int)
        with open(test_y_file, 'r') as fy:
            for idx, (x_line,y_line) in enumerate(itertools.izip(fx,fy)):
                x = get_json_courses(x_line)
                XY[idx, :] = x
                XY[idx, 0] = int(y_line) # Mathematics has index 0
    # split XY in X and Y
    return XY[:,X_columns], XY[:,0]
```

```python
# Read in the training and test data
X_train, Y_train = read_training_data()
X_test, Y_test = read_test_data()

print 'train shape: ', X_train.shape, Y_train.shape
print 'test shape: ', X_test.shape, Y_test.shape
```

```python
# Plot the distribution between each course and mathematics.
X_course_list = [
    'English', # 0
    'Physics', # 1
    'Chemistry', # 2
    'ComputerScience', # 3
    'Biology', # 4
    'PhysicalEducation', # 5
    'Economics', # 6
    'Accountancy', # 7
    'BusinessStudies' # 8
]

for i in xrange(0,9):
    course = X_course_list[i]
    print '\n' + course
    X = X_train[:,i-1]
    Y = Y_train
    # Get the linear correlation
    cor = scipy.stats.pearsonr(X, Y)[0]
    # Plot the 2D distribution of the data
    fig = plt.figure()
    plt.hist2d(X, Y, bins=(8,8))
    ax = plt.gca()
    plt.colorbar()
    plt.ylabel('Mathematics')
    plt.xlabel(X_course_list[i])
    plt.xlim(1, 8)
    plt.ylim(1, 8)
    plt.title("{} vs {} with correlation: {:.2f}".format(X_course_list[i], 'Mathematics', cor))
    plt.show()
```

```python
# Define a function to compute the final score

# p = abs(Predicted Grade Point in Mathematics - Actual Grade Point in Mathematics)
# Where 'abs' indicates the Absolute Value or Magnitude. 
# If p = 0 or 1 your answer for that particular student record will be considered correct. 
# i.e, we allow a tolerance of one grade point away from the correct answer, to take into 
# consideration the marginal errors which might occur during the testing or grading process.

# Score = 100 * ((C-W)/N) 
# Where C = Number of Correct predictions, not more than one grade point away from the 
# actual grade point assigned. 
# W = Number of wrong (incorrect) predictions and 
# N = Total number of records in the input.

def compute_score(target, predicted):
    """
    Function to compute the final score according to the challenge's assignment.
    """
    # make sure both target and predicted are 1D arrays
    if len(target.shape) > 1:
        target = target[:,0]
    if len(predicted.shape) > 1:
        predicted = predicted[:,0]
    if len(target.shape) > 1 or len(predicted.shape) > 1:
        raise ValueError('shape is not 1D. target shape: ' + \
                         str(target.shape) + ', predicted.shape: ' + str(predicted.shape))
    target = np.round(target).astype(int)
    predicted = np.round(predicted).astype(int)
    p = np.abs(predicted - target)
    # count number of correct and incorrect predictions
    counts = np.bincount(p)
    c = np.sum(counts[0:2])
    w = np.sum(counts[2:])
    n = float(target.shape[0])
    # Score = 100 * ((C-W)/N) 
    return 100 * ((c-w) / n)
```

```python
def convert_output(output):
    """
    Function to convert the output to integer values between 2 and 7
    """
    output = np.round(output).astype(int)
    # make sure the output is between 1, and 8.
    # since an error of 1 can be made, the minimum prediction is 2, maximum is 7
    output[output < 2] = 2
    output[output > 7] = 7
    return output
```

```python
# Train a regresser model based on linear support vector machines

def scorer(estimator, X, y):
    """
    Scorer function used during Grid search optimization
    """
    out = estimator.predict(X)
    out = convert_output(out)
    # compute the score
    return compute_score(y, out)


# Tune the parameters of the model with grid search
tuned_parameters = {
    'alpha':np.logspace(-10, 1, 10),
    'epsilon':np.linspace(0, 1, 10)
}
m = sklearn.linear_model.SGDRegressor(
    penalty='l2',
    loss='epsilon_insensitive')
model = sklearn.grid_search.GridSearchCV(m, tuned_parameters, scoring=scorer)
model = model.fit(X_train, Y_train)

print 'Best model parameters found: ', model.best_params_

# Test the final prediction of the model
output = model.predict(X_test)
output = convert_output(output)
# compute the score
print 'score: ', compute_score(Y_test, output)
```

```python
# Train the model on the full dataset that we have available before the final test

# Merge train and test results for training
X = np.concatenate((X_train, X_test), axis=0)
Y = np.concatenate((Y_train, Y_test), axis=0)

# Get best model from GridSearch
best_model = model.best_estimator_ 

# Train on full dataset
best_model = best_model.fit(X, Y)
```

```python
# Save the model to a string so it can be loaded again on hackerrank
file_name = 'model.txt'

# Save the model
serialized = base64.encodestring(zlib.compress(pickle.dumps(best_model),9))
with open(file_name, 'w') as f:
    f.write(serialized)

# open and test the serialization
with open(file_name, 'r') as f:
    serialized = f.read()
    final_model_ff = pickle.loads(zlib.decompress(base64.decodestring(serialized)))

print 'filesize: ', os.path.getsize(file_name)
```
